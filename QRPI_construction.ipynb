```python
amenities_columns = ['Gym', 'SwimmingPool', 'ChildrensPlayArea', 'Jacuzzi', 'Sauna', 'SteamRoom', 'BarbecueArea', 'CCTVSecurity', 'SecurityStaff', 'ConciergeService', 'Intercom', 'AirConditioning', 'CentralHeating', 'Balcony', 'Fireplace', 'WalkInCloset', 'MaidRoom', 'DriversRoom', 'StudyRoom', 'PrivateGarden', 'PrivateJacuzzi', 'PrivatePool', 'PrivateGarage', 'Elevator', 'ElectricityBackup', 'BroadbandInternet', 'SatelliteCableTV', 'WasteDisposal', 'Networked', 'Parking', 'KitchenAppliances', 'LaundryRoom', 'ReceptionWaitingRoom', 'SharedGym', 'SharedSpa', 'PrayerRoom', 'NoCommission', 'PetsAllowed', 'FreeUtilities']
```

```python
non_amen_cols = list(set(list(data.columns)) - set(amenities_columns))
```

```python
# categorical_cols = ['property_type', 'region', 'furnished', 'bedrooms', 'bathrooms']

categorical_cols = ['property_type', 'upd_municip', 'furnished', 'bedrooms', 'bathrooms']
```

```python
data = data.drop(columns=['title', 'link', 'location', 'description',
                          'amenities', 'currency', 'period',
                          'website', 'municipality',
                          'com_type'])
```

```python
data['bathrooms'] = data['bathrooms'].astype(str)
data = data[~data['bathrooms'].str.contains('sqm')]
data['bedrooms'] = data['bedrooms'].astype(str)
data = data[~data['bedrooms'].str.contains('sqm')]
data = data[~data['bedrooms'].str.contains('Viva Bahriya, The Pearl, Doha')]
data = data[~data['bedrooms'].str.contains('Al Mansoura')]
data = data[~data['bedrooms'].str.contains('Furnished')]
```

```python
data
```

```python
data_dummies = pd.get_dummies(data[categorical_cols])
data_dummies = data_dummies.astype(int)
data_model = pd.concat([data, data_dummies], axis=1)
```

```python
data.bathrooms.unique()
```

```python
data_model = data_model.drop(columns=['country', "property_type",  "furnished", 'bedrooms', 'bathrooms', 'cat_bedrooms'])
```

```python
if 'offers' in data_model.columns:
    data_model = data_model.drop(columns=['offers'])
```

```python
data_model['month_year'] = data_model['year'].astype(str) + "_" +data_model['month'].astype(str)
data_model['month_year'].unique()
```

```python
date_dummies = pd.get_dummies(data_model['month_year'])
date_dummies = date_dummies.astype(int)
print(date_dummies)
data_model = pd.concat([data_model, date_dummies], axis=1)
```

```python
data_model = data_model.drop(columns=['year', 'month', 'month_year'])
```

```python
categories = {
    'Health_and_Fitness': [
        'Gym', 'SwimmingPool', 'ChildrensPlayArea', 'Jacuzzi', 'Sauna', 'SteamRoom',
        'SharedGym', 'SharedSpa'
    ],
    'Security_and_Safety': [
        'CCTVSecurity', 'SecurityStaff', 'Intercom'
    ],
    'Home_Comforts_and_Luxuries': [
        'AirConditioning', 'CentralHeating', 'Balcony', 'Fireplace', 'WalkInCloset',
        'MaidRoom', 'DriversRoom', 'StudyRoom', 'PrivateGarden', 'PrivateJacuzzi',
        'PrivatePool', 'PrivateGarage', 'Elevator', 'KitchenAppliances', 'LaundryRoom',
        'ReceptionWaitingRoom', 'PetsAllowed'
    ],
    'Utilities_and_Connectivity': [
        'ElectricityBackup', 'BroadbandInternet', 'SatelliteCableTV', 'WasteDisposal',
        'Networked'
    ],
    'Building_and_Community_Features': [
        'BarbecueArea', 'ConciergeService', 'PrayerRoom', 'NoCommission',
        'FreeUtilities', 'Parking'
    ]
}

# Create new columns for each category
for category, amenities in categories.items():
    # Check if any of the amenities in the category is present for each row
    data_model[category] = data_model[amenities].any(axis=1).astype(int)

amenities_columns = [amenity for sublist in categories.values() for amenity in sublist]
data_model.drop(columns=amenities_columns, inplace=True)
```
