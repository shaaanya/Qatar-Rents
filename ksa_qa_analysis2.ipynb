```python
def find_property_type(row):    
    # List of property types to look for
    if not isinstance(row['type'], float):
        return row['type']
    
    property_types = [
        'Studio', "Apartment", "Villa", "Compound", "Townhouse", "Penthouse", "Duplex",
        "Hotel Apartment", "Whole Building", "Bulk Rent Unit", "Room",
        "Villa Compound", "Residential Building"
    ]

    # Combined text from title and description
    if not isinstance(row['description'], float):
        combined_text = f"{row['title']} {' '.join(row['description'].split()[0:20])}"
    else:
        combined_text = f"{row['title']}"

    # Check for specific combinations and resolve ambiguities
    if "studio apartment" in combined_text.lower():
        return "Studio"
    elif "villa compound" in combined_text.lower():
        return "Villa Compound"
    elif "residential building" in combined_text.lower():
        return "Residential Building"

    # General search for property types
    for p_type in property_types:
        if p_type.lower() in combined_text.lower():
            return p_type

    return "Unsure"
```

```python
def determine_furnished_status(row):
    # Combine title, description, and amenities into one text, handling NaN values
    combined_text = ' '.join(filter(None, [str(row.get(col, '')).lower() for col in ['title', 'description', 'amenities']]))

    # Return None if text is empty
    if not combined_text.strip():
        return None

    # Check for keywords indicating furnished status
    if "not furnished" in combined_text or "unfurnished" in combined_text:
        return "unfurnished"
    elif "partially furnished" in combined_text or "semi furnished" in combined_text or 'partly furnished' in combined_text:
        return "partially_furnished"
    elif "fully furnished" in combined_text or "furnished" in combined_text:
        return "fully_furnished"
    else:
        # If it's not possible to distinguish, return unsure
        return "unsure"

# Applying the function to the DataFrame
data['furnished'] = data.apply(determine_furnished_status, axis=1)
```

```python
def determine_furnished_status(row):
    # Combine title, description, and amenities into one text, handling NaN values
    combined_text = ' '.join(filter(None, [str(row.get(col, '')).lower() for col in ['title', 'description', 'amenities']]))

    # Return None if text is empty
    if not combined_text.strip():
        return None

    # Check for keywords indicating furnished status
    if "not furnished" in combined_text or "unfurnished" in combined_text:
        return "unfurnished"
    elif "partially furnished" in combined_text or "semi furnished" in combined_text or 'partly furnished' in combined_text:
        return "partially_furnished"
    elif "fully furnished" in combined_text or "furnished" in combined_text:
        return "fully_furnished"
    else:
        # If it's not possible to distinguish, return unsure
        return "unsure"

# Applying the function to the DataFrame
data['furnished'] = data.apply(determine_furnished_status, axis=1)
```

```python
def remove_outliers_multiple_categories(df, numeric_column, category_columns, n_std=3):
    grouped = df.groupby(category_columns)

    def remove_outliers(group, n_std=3.5):
        mean = group[numeric_column].mean()
        std_dev = group[numeric_column].std()
        lower_bound = mean - (n_std * std_dev)
        upper_bound = mean + (n_std * std_dev)
        return group[(group[numeric_column] >= lower_bound) & (group[numeric_column] <= upper_bound)]

    filtered_df = grouped.apply(remove_outliers, n_std=n_std).reset_index(drop=True)

    return filtered_df

data_without_outliers = remove_outliers_multiple_categories(data, 'price', ['region', 'property_type'])
new_data_without_outliers = remove_outliers_multiple_categories(new_data, 'price', ['region', 'property_type'])
```
